```python
import cv2
from google.colab import files
import numpy as np
import matplotlib.pyplot as plt

# 1. Upload a photo manually
uploaded = files.upload()

# 2. Read image (uploaded file name)
image_path = list(uploaded.keys())[0]
img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)  # Automatically converts to 8-bit grayscale

# 3. Check matrix shape and values
print("Image shape (HxW):", img.shape)
print("Matrix (first 5 rows):\n", img[:255])

# 4. Show the grayscale image
plt.imshow(img, cmap='gray')
plt.title("Grayscale Image")
plt.axis("off")
plt.show()
```
