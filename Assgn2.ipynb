```python
import numpy as np
import matplotlib.pyplot as plt

# Define image dimensions
width = 256
height = 256

# 1. Create a blank black image using NumPy.
# The shape is (height, width) and the data type is uint8 (0-255).
#using numpy function we fil the array with zeros now do the other thing
A = np.zeros((height, width), dtype=np.uint8)
```

```python
print(A)
```

```python
plt.imshow(A)
```

```python
# 2. Loop through each row to set its intensity.
# This is equivalent to the 'for i = 1 : 256' loop in MATLAB.
# Python is 0-indexed, so we loop from 0 to 255.
for j in range(width):
    # Set all pixels in the current row 'i' to the value 'i'.
    # This is equivalent to A(i, :) = i - 1 in MATLAB.
    A[:, j] = j
```

```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
image = np.zeros((180, 180), dtype=np.uint8)
```

```python
image
```

```python
size=180
width = height = size  # width andd the height of the square
center_x, center_y = width // 2, height // 2
```

```python
radius = size // 2
```

```python
for i in range(height):  # Corresponds to y-coordinate (row)
    for j in range(width): # Corresponds to x-coordinate (column)

        # Calculate the distance of the pixel (j, i) from the center.
            distance = math.sqrt((i - center_y)**2 + (j - center_x)**2)

            # If the distance is less than the radius, it's inside the circle.
            if distance < radius:
                image[i, j] = 255
```

```python
plt.imshow(image, cmap='gray', vmin=0, vmax=255)
```

```python
def manual_transpose(image):
    """Transposes an image by swapping row and column indices."""
    # Get the height and width of the original image
    height, width = image.shape

    # Create a new blank image with swapped dimensions
    transposed_img = np.zeros((width, height), dtype=np.uint8)

    # Loop through each pixel of the original image
    for i in range(height):
        for j in range(width):
            # Copy the pixel from (i, j) to (j, i) in the new image
            transposed_img[j, i] = image[i, j]

    return transposed_img





```

```python
ans=manual_transpose(img)
```

```python
plt.imshow(ans,cmap='gray')
```

```python
def manual_flip_vertical(image):
    """Flips an image vertically (upside down)."""
    height, width = image.shape
    flipped_img = np.zeros((height, width), dtype=np.uint8)

    for i in range(height):
        for j in range(width):
            # Copy pixel from row 'i' to the new row 'height - 1 - i'
            flipped_img[height - 1 - i, j] = image[i, j]

    return flipped_img
```

```python
flp1=manual_flip_vertical(img)
```

```python
plt.imshow(flp1,cmap='gray')
```

```python
def manual_flip_horizontal(image):
    """Flips an image vertically (upside down)."""
    height, width = image.shape
    flipped_img = np.zeros((height, width), dtype=np.uint8)

    for i in range(height):
        for j in range(width):
            # Copy pixel from row 'i' to the new row 'height - 1 - i'
            flipped_img[i, width-1-j] = image[i, j]

    return flipped_img
```

```python
flp2=manual_flip_horizontal(img)
```

```python
plt.imshow(flp2,cmap='gray')
```

```python
def manual_flip_diagonal(image):
    """Flips an image vertically (upside down)."""
    height, width = image.shape
    flipped_img = np.zeros((height, width), dtype=np.uint8)

    for i in range(height):
        for j in range(width):
            # Copy pixel from row 'i' to the new row 'height - 1 - i'
            flipped_img[height - 1 - i, width-1-j] = image[i, j]

    return flipped_img
```

```python
flp3=manual_flip_diagonal(img)
```

```python
plt.imshow(flp3,cmap='gray')
```
